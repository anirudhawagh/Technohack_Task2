# **Problem Statement** - calculate summary statistics(mean , median , mode, standard derivation ) for a dataset

```python
# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
print("Categorical columns:")
print(categorical_columns)
```

```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Numerical columns:")
print(numerical_columns)
```

```python
# Select numerical columns
numerical_columns = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Calculate mean, median, mode, and standard deviation for each numerical column
summary_statistics = pd.DataFrame({
   # 'Column': numerical_columns,
    'Mean': df[numerical_columns].mean(),
    'Median': df[numerical_columns].median(),
    'Mode': [mode(df[col])[0] for col in numerical_columns],
    'Standard Deviation': df[numerical_columns].std()
})

# Display the summary statistics table
summary_statistics
```

```python
# Calculate summary statistics for 'PassengerId' grouped by 'Sex'
passenger_id_statistics = df.groupby('Sex')['PassengerId'].agg(['mean', 'median', lambda x: mode(x)[0], 'std']).reset_index()

# Rename the lambda function column to 'Mode'
passenger_id_statistics = passenger_id_statistics.rename(columns={'<lambda_0>': 'Mode'})

# Create a new DataFrame to format the table
formatted_passenger_id_statistics = pd.DataFrame({
    'Sex': ['Female', 'Male'],
    'Mean': passenger_id_statistics['mean'],
    'Median': passenger_id_statistics['median'],
    'Mode': passenger_id_statistics['Mode'],
    'Standard Deviation': passenger_id_statistics['std']
})

# Add a line indicating that this table represents 'PassengerId' statistics
print("PassengerId Statistics:")
# Display the formatted summary statistics table for 'PassengerId'
print(formatted_passenger_id_statistics)
```
